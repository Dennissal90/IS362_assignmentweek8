## Objective
In this assignment, I will load and analyze the Auto MPG dataset from the UC Irvine repository. The goal is to perform basic data cleanup, transformations, and conduct exploratory data analysis (EDA) to understand the distribution of variables and the relationships between them.

### Step 1: Import Libraries
I'll start by importing the necessary libraries for data manipulation and visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

### Step 2: Load the Data
I'll load the dataset from the provided file and convert it into a Pandas DataFrame.

```python
file_path = r"C:\Users\The King\Desktop\FALL 2024\IS 362\IS362_Week8_assignment\auto-mpg.data"
column_names = [
    "mpg", "cylinders", "displacement", "horsepower", "weight", "acceleration", "model_year", "origin", "car_name"
]
df_auto_mpg = pd.read_csv(file_path, delim_whitespace=True, names=column_names)
print(df_auto_mpg.head())
```

### Step 3: Data Cleaning and Transformation
1. Replace missing values in the `horsepower` column.
2. Convert the `origin` column to more descriptive labels.

```python
# Replace '?' in horsepower with NaN, then convert to numeric
df_auto_mpg['horsepower'] = df_auto_mpg['horsepower'].replace('?', np.nan).astype('float')

# Convert origin column to categorical values
df_auto_mpg['origin'] = df_auto_mpg['origin'].replace({1: 'USA', 2: 'Asia', 3: 'Europe'})
print(df_auto_mpg.info())
```

### Step 4: Exploratory Data Analysis (EDA)
I'll start by visualizing the distribution of cylinders in the dataset.

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='cylinders', data=df_auto_mpg, palette='viridis')
plt.title('Distribution of Cylinders')
plt.xlabel('Number of Cylinders')
plt.ylabel('Count')
plt.show()
```

#### Step 4.2: Scatter Plot of Horsepower vs Weight
I'll visualize the relationship between `horsepower` and `weight` to see if there are any noticeable trends.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='horsepower', y='weight', data=df_auto_mpg, hue='cylinders', palette='viridis')
plt.title('Horsepower vs. Weight')
plt.xlabel('Horsepower')
plt.ylabel('Weight')
plt.show()
```

#### Step 4.3: Exploring an Interesting Question
I'll ask an interesting question about the data: "What is the distribution of fuel efficiency (`mpg`) for cars made in different regions?"

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='origin', y='mpg', data=df_auto_mpg, palette='viridis')
plt.title('Distribution of MPG by Origin')
plt.xlabel('Region of Origin')
plt.ylabel('Miles per Gallon (MPG)')
plt.show()
```

### Conclusion
This assignment demonstrates how to load and clean a dataset, perform basic transformations, and conduct exploratory analysis using visualizations. The EDA reveals insights about the distribution of car characteristics, such as the distribution of cylinders, the relationship between horsepower and weight, and how fuel efficiency differs by region of origin.
